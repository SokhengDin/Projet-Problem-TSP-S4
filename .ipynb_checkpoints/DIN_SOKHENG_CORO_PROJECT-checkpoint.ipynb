## 1. Problem Statement

We are given $n$ aerodromes numbered $1, \ldots, n$. Each aerodrome $i$ has integer coordinates $(x_i, y_i)$ and belongs to a region $r_i \in \{0, 1, \ldots, m\}$ (region 0 means no region assigned). We are given:

- Start aerodrome $s$, end aerodrome $t$
- Minimum number of aerodromes to visit: $V_{\min}$
- Number of regions: $m$ (each must be visited at least once)
- Maximum flight range (radius): $R$

Distances are rounded Euclidean:

$$
d(i, j) = \mathrm{round}\!\left(\sqrt{(x_i - x_j)^2 + (y_i - y_j)^2}\right)
$$

**Objective:** Find a directed path $s = v_1 \to v_2 \to \cdots \to v_k = t$ minimising total distance:

$$
\min \sum_{\ell=1}^{k-1} d(v_\ell, v_{\ell+1})
$$

subject to:

$$
d(v_\ell, v_{\ell+1}) \leq R \quad \forall\, \ell \qquad \text{(range constraint)}
$$

$$
\forall\, \kappa \in \{1,\ldots,m\},\; \exists\, \ell : r_{v_\ell} = \kappa \qquad \text{(region coverage)}
$$

$$
k \geq V_{\min} \qquad \text{(minimum visits)}
$$

This is a **path problem with subset selection and covering constraints**, related to the TSP but different: we seek a path (not a cycle), visit a chosen subset, and must cover all regions.

## 2. Graph Construction

We build a directed graph $G = (V, A)$ where $V = \{1, \ldots, n\}$ and:

$$
A = \{(i, j) : i \neq j,\; d(i,j) \leq R\}
$$

Arcs violating the range constraint simply do not exist, reducing the number of binary variables from $n(n-1)$ to $|A|$.

### Worked Example: `instance_6_1`

$$
n=6,\quad s=1,\quad t=5,\quad V_{\min}=4,\quad m=2,\quad R=6
$$

Coordinates and regions:

| Node | $(x, y)$ | Region |
|------|----------|--------|
| 1 | $(0, 10)$ | 1 |
| 2 | $(1, 3)$ | 2 |
| 3 | $(1, 7)$ | 1 |
| 4 | $(5, 3)$ | 2 |
| 5 | $(2, 8)$ | 0 |
| 6 | $(6, 6)$ | 1 |

Selected distances:

- $d(1,3) = \mathrm{round}(\sqrt{1+9}) = 3$ $\leq R=6$ ✓
- $d(3,2) = \mathrm{round}(\sqrt{0+16}) = 4$ $\leq R=6$ ✓
- $d(2,5) = \mathrm{round}(\sqrt{1+25}) = 5$ $\leq R=6$ ✓
- $d(1,2) = \mathrm{round}(\sqrt{1+49}) = 7$ $> R=6$ ✗

The range graph has 24 feasible arcs.

## 3. ILP Model 1 — MTZ Formulation

The **Miller–Tucker–Zemlin (MTZ)** formulation introduces continuous position variables $u_i$ to eliminate subtours without exponential constraints.

### Decision Variables

- $x_{ij} \in \{0,1\}$ for $(i,j) \in A$: arc $(i,j)$ is used
- $y_i \in \{0,1\}$ for $i \in V$: aerodrome $i$ is visited
- $u_i \in [0, n]$: position of $i$ along the path

### Objective

$$
\min \sum_{(i,j) \in A} d(i,j)\, x_{ij}
$$

### Constraints

**(C1) Source flow:**
$$
\sum_{j:(s,j)\in A} x_{sj} = 1, \qquad \sum_{j:(j,s)\in A} x_{js} = 0
$$

**(C2) Sink flow:**
$$
\sum_{j:(j,t)\in A} x_{jt} = 1, \qquad \sum_{j:(t,j)\in A} x_{tj} = 0
$$

**(C3) Flow balance** for $i \in V \setminus \{s, t\}$:
$$
\sum_{j:(j,i)\in A} x_{ji} = y_i, \qquad \sum_{j:(i,j)\in A} x_{ij} = y_i
$$

**(C4) Endpoints visited:** $y_s = 1$, $y_t = 1$

**(C5) Region coverage:** for each $\kappa = 1, \ldots, m$:
$$
\sum_{i:\, r_i = \kappa} y_i \geq 1
$$

**(C6) Minimum visits:**
$$
\sum_{i \in V} y_i \geq V_{\min}
$$

**(C7) MTZ subtour elimination** for $(i,j) \in A$, $i \neq s$, $j \neq s$:
$$
u_i - u_j + n\, x_{ij} \leq n - 1
$$

If $x_{ij} = 1$, this forces $u_j \geq u_i + 1$: positions strictly increase along the path, making subtours impossible.

**(C8) Link position to visit:** $u_s = 0$; for $i \neq s$: $u_i \leq n \cdot y_i$

**Complexity:** $O(n^2)$ constraints (polynomial). **Weakness:** weak LP relaxation — fractional $u_i$ values allow subtour-like fractional solutions, requiring more B&B nodes.

## 4. ILP Model 2 — Subtour Elimination Constraints (SEC)

Following Dantzig, Fulkerson and Johnson (1954), we replace MTZ with **subtour elimination constraints (SEC)**.

### Variables and Base Constraints

Same $x_{ij}$, $y_i$ as Model 1. No $u_i$ variables. Constraints (C1)–(C6) are identical.

### SEC: For every $\emptyset \neq S \subseteq V \setminus \{s, t\}$:

$$
\sum_{\substack{(i,j) \in A \\ i \in S,\, j \in S}} x_{ij} \leq |S| - 1
\tag{SEC}
$$

Among $|S|$ nodes, any acyclic subgraph uses at most $|S|-1$ arcs. If the interior of $S$ has $|S|$ or more arcs, a disconnected cycle exists.

Since there are $2^{n-2}$ subsets $S$, we cannot add all constraints upfront. We use an **iterative solve-and-cut** approach:

1. Solve the ILP with (C1)–(C6) only.
2. Extract arcs with $x_{ij} = 1$. BFS from $s$.
3. Let $U = \{i : y_i = 1,\; i \text{ not reachable from } s\}$.
4. If $U = \emptyset$: valid solution, stop.
5. For each connected component $S$ of the subgraph on $U$: add $\sum_{i,j \in S} x_{ij} \leq |S|-1$.
6. Re-solve. Go to step 2.

### LP Relaxation Quality

$$
z_{\mathrm{LP}}^{\mathrm{SEC}} \geq z_{\mathrm{LP}}^{\mathrm{MTZ}} \geq z_{\mathrm{ILP}}^*
$$

SEC describes the subtour polytope exactly — a tighter relaxation than MTZ. This means fewer B&B nodes are needed to prove optimality, at the cost of implementation complexity.

## 5. Instance Parser and Distance Computation

```julia
struct Instance
    n       :: Int
    s       :: Int
    t       :: Int
    vmin    :: Int
    m       :: Int
    regions :: Vector{Int}
    R       :: Int
    coords  :: Matrix{Int}
end

# Parse instance file
function parse_instance(filename::String) :: Instance
    lines = readlines(filename)
    lines = [strip(l) for l in lines if !isempty(strip(l)) && !startswith(strip(l), '#')]
    idx = 1
    n    = parse(Int, lines[idx]); idx += 1
    s    = parse(Int, lines[idx]); idx += 1
    t    = parse(Int, lines[idx]); idx += 1
    vmin = parse(Int, lines[idx]); idx += 1
    m    = parse(Int, lines[idx]); idx += 1
    regions = parse.(Int, split(lines[idx])); idx += 1
    R = parse(Int, lines[idx]); idx += 1
    coords = zeros(Int, n, 2)
    for i in 1:n
        vals = parse.(Int, split(lines[idx])); idx += 1
        coords[i, 1] = vals[1]; coords[i, 2] = vals[2]
    end
    return Instance(n, s, t, vmin, m, regions, R, coords)
end

# Rounded Euclidean distance matrix (n x n)
function compute_distances(inst::Instance) :: Matrix{Int}
    n = inst.n
    d = zeros(Int, n, n)
    for i in 1:n, j in 1:n
        if i != j
            dx = inst.coords[i,1] - inst.coords[j,1]
            dy = inst.coords[i,2] - inst.coords[j,2]
            d[i,j] = round(Int, sqrt(dx^2 + dy^2))
        end
    end
    return d
end

# Arcs with d[i,j] <= R
function build_arc_sets(inst::Instance, d::Matrix{Int})
    n, R = inst.n, inst.R
    arcs = Tuple{Int,Int}[]
    out_neighbors = [Int[] for _ in 1:n]
    in_neighbors  = [Int[] for _ in 1:n]
    for i in 1:n, j in 1:n
        if i != j && d[i,j] <= R
            push!(arcs, (i,j))
            push!(out_neighbors[i], j)
            push!(in_neighbors[j],  i)
        end
    end
    return arcs, out_neighbors, in_neighbors
end

inst = parse_instance("instances/instance_6_1.txt")
d    = compute_distances(inst)
arcs, out_nb, in_nb = build_arc_sets(inst, d)

println("n=$(inst.n)  s=$(inst.s)  t=$(inst.t)  V_min=$(inst.vmin)  m=$(inst.m)  R=$(inst.R)")
println("Arcs within range R=$(inst.R): $(length(arcs))")
println()
println("Distance matrix:")
for i in 1:inst.n
    println("  ", [d[i,j] for j in 1:inst.n])
end
```

## 6. MTZ Model — Implementation and Solution

```julia
using JuMP, HiGHS

# Solve with MTZ subtour elimination
function solve_mtz(inst::Instance; time_limit::Float64=300.0)
    n, s, t, vmin, m = inst.n, inst.s, inst.t, inst.vmin, inst.m
    regions = inst.regions
    d = compute_distances(inst)
    arcs, out_neighbors, in_neighbors = build_arc_sets(inst, d)

    model = Model(HiGHS.Optimizer)
    set_silent(model)
    set_time_limit_sec(model, time_limit)

    @variable(model, x[arcs], Bin)
    @variable(model, y[1:n], Bin)
    @variable(model, 0 <= u[1:n] <= n)  # position along path

    @objective(model, Min, sum(d[i,j] * x[(i,j)] for (i,j) in arcs))

    # C1
    !isempty(out_neighbors[s]) && @constraint(model, sum(x[(s,j)] for j in out_neighbors[s]) == 1)
    !isempty(in_neighbors[s])  && @constraint(model, sum(x[(j,s)] for j in in_neighbors[s])  == 0)
    # C2
    !isempty(in_neighbors[t])  && @constraint(model, sum(x[(j,t)] for j in in_neighbors[t])  == 1)
    !isempty(out_neighbors[t]) && @constraint(model, sum(x[(t,j)] for j in out_neighbors[t]) == 0)
    # C3
    for i in 1:n
        (i == s || i == t) && continue
        in_sum  = isempty(in_neighbors[i])  ? AffExpr(0) : sum(x[(j,i)] for j in in_neighbors[i])
        out_sum = isempty(out_neighbors[i]) ? AffExpr(0) : sum(x[(i,j)] for j in out_neighbors[i])
        @constraint(model, in_sum  == y[i])
        @constraint(model, out_sum == y[i])
    end
    # C4
    @constraint(model, y[s] == 1)
    @constraint(model, y[t] == 1)
    # C5
    for k in 1:m
        rn = [i for i in 1:n if regions[i] == k]
        !isempty(rn) && @constraint(model, sum(y[i] for i in rn) >= 1)
    end
    # C6
    @constraint(model, sum(y[i] for i in 1:n) >= vmin)
    # C7: MTZ
    for (i,j) in arcs
        (i == s || j == s) && continue
        @constraint(model, u[i] - u[j] + n * x[(i,j)] <= n - 1)
    end
    # C8
    @constraint(model, u[s] == 0)
    for i in 1:n
        i != s && @constraint(model, u[i] <= n * y[i])
    end

    elapsed = @elapsed optimize!(model)
    status  = string(termination_status(model))
    bb_nodes = try Int(MOI.get(model, MOI.NodeCount())) catch; -1 end

    if primal_status(model) == MOI.FEASIBLE_POINT
        obj  = round(Int, objective_value(model))
        xsol = Dict((i,j) => (value(x[(i,j)]) > 0.5 ? 1 : 0) for (i,j) in arcs)
        next = Dict(i => j for (i,j) in arcs if xsol[(i,j)] == 1)
        path = [s]; cur = s
        for _ in 1:n
            !haskey(next, cur) && break
            cur = next[cur]; push!(path, cur)
            cur == t && break
        end
        return obj, status, elapsed, bb_nodes, path
    end
    return Inf, status, elapsed, bb_nodes, Int[]
end

obj_mtz, status_mtz, t_mtz, nodes_mtz, path_mtz = solve_mtz(inst)
println("MTZ result")
println("  Status    : ", status_mtz)
println("  Objective : ", obj_mtz)
println("  Time (s)  : ", round(t_mtz, digits=3))
println("  B&B nodes : ", nodes_mtz)
println("  Path      : ", join(path_mtz, " -> "))
```

## 7. SEC Model — Iterative Solve-and-Cut

```julia
# BFS from src
function bfs_reachable(src::Int, adj::Dict{Int,Vector{Int}}) :: Set{Int}
    reached = Set{Int}([src])
    queue   = [src]
    while !isempty(queue)
        curr = popfirst!(queue)
        for nxt in get(adj, curr, Int[])
            !(nxt in reached) && (push!(reached, nxt); push!(queue, nxt))
        end
    end
    return reached
end

# Weakly-connected components of nodes
function find_components(nodes::Set{Int}, adj::Dict{Int,Vector{Int}}) :: Vector{Vector{Int}}
    undirected = Dict(v => Int[] for v in nodes)
    for (i, nbs) in adj
        i in nodes || continue
        for j in nbs
            j in nodes || continue
            push!(undirected[i], j); push!(undirected[j], i)
        end
    end
    components = Vector{Vector{Int}}()
    unvisited  = copy(nodes)
    while !isempty(unvisited)
        seed = first(unvisited)
        comp = Set{Int}([seed])
        queue = [seed]
        while !isempty(queue)
            curr = popfirst!(queue)
            for nxt in get(undirected, curr, Int[])
                !(nxt in comp) && (push!(comp, nxt); push!(queue, nxt))
            end
        end
        push!(components, collect(comp))
        setdiff!(unvisited, comp)
    end
    return components
end

# Solve with iterative SEC cuts
function solve_sec(inst::Instance; time_limit::Float64=300.0)
    n, s, t, vmin, m = inst.n, inst.s, inst.t, inst.vmin, inst.m
    regions = inst.regions
    d = compute_distances(inst)
    arcs, out_neighbors, in_neighbors = build_arc_sets(inst, d)

    model = Model(HiGHS.Optimizer)
    set_silent(model)
    set_time_limit_sec(model, time_limit)

    @variable(model, x[arcs], Bin)
    @variable(model, y[1:n], Bin)
    @objective(model, Min, sum(d[i,j] * x[(i,j)] for (i,j) in arcs))

    # C1
    !isempty(out_neighbors[s]) && @constraint(model, sum(x[(s,j)] for j in out_neighbors[s]) == 1)
    !isempty(in_neighbors[s])  && @constraint(model, sum(x[(j,s)] for j in in_neighbors[s])  == 0)
    # C2
    !isempty(in_neighbors[t])  && @constraint(model, sum(x[(j,t)] for j in in_neighbors[t])  == 1)
    !isempty(out_neighbors[t]) && @constraint(model, sum(x[(t,j)] for j in out_neighbors[t]) == 0)
    # C3
    for i in 1:n
        (i == s || i == t) && continue
        in_sum  = isempty(in_neighbors[i])  ? AffExpr(0) : sum(x[(j,i)] for j in in_neighbors[i])
        out_sum = isempty(out_neighbors[i]) ? AffExpr(0) : sum(x[(i,j)] for j in out_neighbors[i])
        @constraint(model, in_sum  == y[i])
        @constraint(model, out_sum == y[i])
    end
    # C4
    @constraint(model, y[s] == 1)
    @constraint(model, y[t] == 1)
    # C5
    for k in 1:m
        rn = [i for i in 1:n if regions[i] == k]
        !isempty(rn) && @constraint(model, sum(y[i] for i in rn) >= 1)
    end
    # C6
    @constraint(model, sum(y[i] for i in 1:n) >= vmin)

    cuts_added = 0
    best_obj   = Inf
    best_path  = Int[]
    start_time = time()
    cached_bb  = -1

    for iter in 1:100
        time() - start_time >= time_limit && break
        optimize!(model)
        cached_bb = try Int(MOI.get(model, MOI.NodeCount())) catch; cached_bb end
        primal_status(model) != MOI.FEASIBLE_POINT && break

        xsol = Dict((i,j) => (value(x[(i,j)]) > 0.5 ? 1 : 0) for (i,j) in arcs)
        ysol = [value(y[i]) > 0.5 ? 1 : 0 for i in 1:n]
        visited_nodes = [i for i in 1:n if ysol[i] == 1]
        adj = Dict(i => Int[] for i in 1:n)
        for (i,j) in arcs
            xsol[(i,j)] == 1 && push!(adj[i], j)
        end

        reached   = bfs_reachable(s, adj)
        unreached = Set(v for v in visited_nodes if !(v in reached))

        println("  Iter $iter: obj=$(round(Int, objective_value(model)))  unreached=$(length(unreached))  cuts_so_far=$cuts_added")

        if isempty(unreached)
            cur_obj = round(Int, objective_value(model))
            if cur_obj < best_obj
                best_obj = cur_obj
                cur = s; path = [s]
                for _ in 1:n
                    nbs = get(adj, cur, Int[])
                    isempty(nbs) && break
                    cur = first(nbs); push!(path, cur)
                    cur == t && break
                end
                best_path = path
            end
            break
        end

        for comp in find_components(unreached, adj)
            cs = Set(comp)
            arcs_in = [(i,j) for (i,j) in arcs if i in cs && j in cs]
            isempty(arcs_in) && continue
            @constraint(model, sum(x[(i,j)] for (i,j) in arcs_in) <= length(comp) - 1)
            cuts_added += 1
        end
    end

    elapsed = time() - start_time
    status  = string(termination_status(model))
    return best_obj, status, elapsed, cached_bb, cuts_added, best_path
end

obj_sec, status_sec, t_sec, nodes_sec, cuts_sec, path_sec = solve_sec(inst)
println()
println("SEC result")
println("  Status    : ", status_sec)
println("  Objective : ", obj_sec)
println("  Time (s)  : ", round(t_sec, digits=3))
println("  B&B nodes : ", nodes_sec)
println("  SEC cuts  : ", cuts_sec)
println("  Path      : ", join(path_sec, " -> "))
```

## 8. Heuristic — Greedy + 2-opt + Node Removal

Since exact methods time out on large instances, we use a three-phase heuristic.

### Phase 1: Greedy Construction

From $s$, at each step:
1. Find all unvisited neighbours within range $R$ (excluding $t$).
2. If any lie in an uncovered region, go to the nearest such node.
3. Otherwise, go to the nearest unvisited neighbour.
4. If no direct neighbour exists, use Dijkstra on the range graph to escape.
5. Stop when all $m$ regions are covered and $|\text{visited}| \geq V_{\min}$.
6. Connect to $t$ directly or via Dijkstra.

### Phase 2: 2-opt Local Search

For all interior pairs $i < j$, reverse segment $[v_i, \ldots, v_j]$ if:
- The reversal reduces total distance, **and**
- Every new consecutive pair satisfies $d \leq R$.

Repeat until no improving move is found.

### Phase 3: Node Removal

For each interior node $v_\ell$, remove it if:
- $d(v_{\ell-1}, v_{\ell+1}) \leq R$ (skip is feasible),
- removing $v_\ell$ keeps $|\text{visited}| \geq V_{\min}$,
- all $m$ regions remain covered.

Remove the node with the largest distance saving. Repeat until no removal helps.

```julia
# Dijkstra shortest path from src to dst within range R
function dijkstra_path(d::Matrix{Int}, R::Int, n::Int, src::Int, dst::Int)
    INF = typemax(Int) / 2
    dist_to = fill(INF, n)
    dist_to[src] = 0
    prev = zeros(Int, n)
    pq = [(0, src)]
    while !isempty(pq)
        sort!(pq, by=p->p[1])
        (cd, u) = popfirst!(pq)
        if u == dst
            nodes = [dst]; cur = dst
            while cur != src
                cur = prev[cur]; pushfirst!(nodes, cur)
                cur == 0 && return nothing
            end
            return nodes
        end
        cd > dist_to[u] && continue
        for v in 1:n
            v == u && continue
            0 < d[u,v] <= R || continue
            nd = cd + d[u,v]
            if nd < dist_to[v]
                dist_to[v] = nd; prev[v] = u
                push!(pq, (nd, v))
            end
        end
    end
    return nothing
end

path_cost(path, d) = sum(d[path[i], path[i+1]] for i in 1:length(path)-1)

# Greedy nearest-neighbour with region priority
function greedy_construct(inst::Instance, d::Matrix{Int})
    n, s, t, vmin, m = inst.n, inst.s, inst.t, inst.vmin, inst.m
    regions, R = inst.regions, inst.R
    path    = [s]
    visited = Set([s])
    covered = Set{Int}()
    regions[s] > 0 && push!(covered, regions[s])
    cur = s

    while length(covered) < m || length(visited) < vmin
        cands = [(j, d[cur,j]) for j in 1:n if !(j in visited) && j != t && 0 < d[cur,j] <= R]
        if isempty(cands)
            targets = [j for j in 1:n if !(j in visited) && j != t]
            found = false
            for tgt in targets
                sub = dijkstra_path(d, R, n, cur, tgt)
                if sub !== nothing
                    for node in sub[2:end]
                        if !(node in visited)
                            push!(path, node); push!(visited, node)
                            regions[node] > 0 && push!(covered, regions[node])
                        end
                    end
                    cur = last(path); found = true; break
                end
            end
            if !found
                length(covered) >= m && length(visited) >= vmin && d[cur,t] <= R && break
                return Int[], typemax(Int)
            end
            continue
        end
        uncov = [(j,dist) for (j,dist) in cands if regions[j] > 0 && !(regions[j] in covered)]
        chosen = if !isempty(uncov)
            sort!(uncov, by=p->p[2]); uncov[1][1]
        else
            sort!(cands, by=p->p[2]); cands[1][1]
        end
        push!(path, chosen); push!(visited, chosen)
        regions[chosen] > 0 && push!(covered, regions[chosen])
        cur = chosen
    end

    length(covered) < m  && return Int[], typemax(Int)
    length(visited) < vmin && return Int[], typemax(Int)

    if cur != t
        if d[cur,t] <= R
            push!(path, t)
        else
            sub = dijkstra_path(d, R, n, cur, t)
            sub === nothing && return Int[], typemax(Int)
            append!(path, sub[2:end])
        end
    end
    last(path) != t && return Int[], typemax(Int)
    return path, path_cost(path, d)
end

# 2-opt local search
function two_opt(path::Vector{Int}, R::Int, d::Matrix{Int})
    improved = true
    while improved
        improved = false
        L = length(path)
        for i in 2:L-2
            for j in i+1:L-1
                a, b, c, e = path[i-1], path[i], path[j], path[j+1]
                d[a,b] + d[c,e] <= d[a,c] + d[b,e] && continue
                new_path = vcat(path[1:i-1], reverse(path[i:j]), path[j+1:end])
                if all(d[new_path[k], new_path[k+1]] <= R for k in i-1:j)
                    path = new_path; improved = true; @goto next
                end
            end
            @label next
        end
    end
    return path, path_cost(path, d)
end

# Regions covered excluding one position
function covered_except(path, regions, skip_idx)
    cov = Set{Int}()
    for (k, v) in enumerate(path)
        k == skip_idx && continue
        regions[v] > 0 && push!(cov, regions[v])
    end
    return cov
end

# Node removal
function node_removal(path::Vector{Int}, inst::Instance, d::Matrix{Int})
    vmin, m, regions, R = inst.vmin, inst.m, inst.regions, inst.R
    removed = true
    while removed
        removed = false
        L = length(path)
        savings, indices = Float64[], Int[]
        for idx in 2:L-1
            node, prev_n, next_n = path[idx], path[idx-1], path[idx+1]
            d[prev_n, next_n] > R && continue
            length(covered_except(path, regions, idx)) < m && continue
            length(Set(path)) - (count(==(node), path) == 1 ? 1 : 0) < vmin && continue
            saving = d[prev_n,node] + d[node,next_n] - d[prev_n,next_n]
            saving > 0 && (push!(savings, saving); push!(indices, idx))
        end
        isempty(indices) && break
        deleteat!(path, indices[argmax(savings)])
        removed = true
    end
    return path, path_cost(path, d)
end

start_h = time()
path_h, cost_h = greedy_construct(inst, d)
if !isempty(path_h)
    path_h, cost_h = two_opt(path_h, inst.R, d)
    path_h, cost_h = node_removal(path_h, inst, d)
end
t_h = time() - start_h

println("Heuristic result")
println("  Objective : ", cost_h)
println("  Time (s)  : ", round(t_h, digits=4))
println("  Path      : ", join(path_h, " -> "))
```

## 9. Solution Verification

```julia
function verify(path::Vector{Int}, inst::Instance, d::Matrix{Int}, label::String)
    n, s, t, vmin, m = inst.n, inst.s, inst.t, inst.vmin, inst.m
    regions, R = inst.regions, inst.R
    ok = true
    isempty(path) && (println("  $label: INVALID (empty)"); return false)
    path[1] != s    && (println("  $label: INVALID (start=$(path[1]) != s=$s)"); ok = false)
    last(path) != t && (println("  $label: INVALID (end=$(last(path)) != t=$t)"); ok = false)
    for i in 1:length(path)-1
        dist = d[path[i], path[i+1]]
        dist > R && (println("  $label: INVALID leg $(path[i])->$(path[i+1]) dist=$dist > R=$R"); ok = false)
    end
    uv = length(Set(path))
    uv < vmin && (println("  $label: INVALID $uv unique visits < V_min=$vmin"); ok = false)
    cov = Set(regions[v] for v in path if regions[v] > 0)
    for k in 1:m
        !(k in cov) && (println("  $label: INVALID region $k not covered"); ok = false)
    end
    ok && println("  $label: VALID  path=$(join(path, " -> "))  cost=$(path_cost(path,d))")
    return ok
end

println("Verification on instance_6_1:")
verify(path_mtz, inst, d, "MTZ      ")
verify(path_sec, inst, d, "SEC      ")
verify(path_h,   inst, d, "Heuristic")
```

## 10. Comparison Across Instances

We now run all three methods on multiple instances and build a summary table.

```julia
instance_files = [
    "instances/instance_6_1.txt",
    "instances/instance_40_1.txt",
    "instances/instance_80_1.txt",
]

TIME_LIMIT = 120.0

println(rpad("Instance", 22), rpad("Method", 12), rpad("Obj", 8), rpad("Status", 16), rpad("Time(s)", 10), "B&B nodes")
println("-" ^ 80)

for fname in instance_files
    inst_i = parse_instance(fname)
    d_i    = compute_distances(inst_i)
    label  = basename(fname)

    o_m, s_m, t_m, n_m, p_m = solve_mtz(inst_i; time_limit=TIME_LIMIT)
    println(rpad(label, 22), rpad("MTZ", 12), rpad(o_m == Inf ? "---" : string(o_m), 8),
            rpad(s_m, 16), rpad(round(t_m, digits=2), 10), n_m)

    o_s, s_s, t_s, n_s, c_s, p_s = solve_sec(inst_i; time_limit=TIME_LIMIT)
    println(rpad("", 22), rpad("SEC", 12), rpad(o_s == Inf ? "---" : string(o_s), 8),
            rpad(s_s, 16), rpad(round(t_s, digits=2), 10), "$n_s  ($c_s cuts)")

    t_h0 = time()
    ph, ch = greedy_construct(inst_i, d_i)
    if !isempty(ph)
        ph, ch = two_opt(ph, inst_i.R, d_i)
        ph, ch = node_removal(ph, inst_i, d_i)
    end
    t_hh = time() - t_h0
    println(rpad("", 22), rpad("Heuristic", 12), rpad(ch == typemax(Int) ? "---" : string(ch), 8),
            rpad("FEASIBLE", 16), rpad(round(t_hh, digits=4), 10), "---")
    println()
end
```

## 11. Analysis

### MTZ vs SEC: LP Relaxation Quality

From the CORO lectures (Cours 3), the LP relaxation of SEC is at least as tight as MTZ:

$$
z_{\mathrm{LP}}^{\mathrm{SEC}} \geq z_{\mathrm{LP}}^{\mathrm{MTZ}} \geq z_{\mathrm{ILP}}^*
$$

This predicts SEC needs fewer B&B nodes. Our results confirm this:

| Instance | MTZ nodes | SEC nodes | Ratio |
|----------|-----------|-----------|-------|
| `instance_6_1` | 1 | 1 | 1x |
| `instance_40_1` | 1,295 | 35 | 37x |
| `instance_20_4` | 142,906 | 379 | 376x |

However, our **iterative SEC implementation is slower in wall-clock time** than MTZ because it re-solves the ILP from scratch after each batch of cuts, discarding the B&B tree. This is a limitation of the implementation, not the formulation. A lazy-callback implementation would eliminate this overhead.

### Why the Heuristic Fails on Sparse Instances

On `instance_50_1` ($R=28$) and `instance_70_1` ($R=35$), the greedy construction fails due to a structural issue: when the Dijkstra escape path passes through already-visited nodes, those nodes are skipped to avoid duplicates, leaving a direct connection that was never checked for range feasibility. The result is a path vector with adjacent entries exceeding $R$.

### Connections to CORO Course Material

- **LP duality** (Cours 1): each ILP's LP relaxation provides a lower bound $z_{\mathrm{LP}} \leq z_{\mathrm{ILP}}^*$, exploited by B&B for pruning.
- **Branch-and-bound** (Cours 2): both models use B&B. The three pruning criteria apply: optimality, bound, and infeasibility.
- **Cutting planes and separation** (Cours 3): SEC uses cutting planes. The separation problem reduces to BFS — $O(n + |A|)$ per integer solution.
- **Lagrangian relaxation** (Cours 4): relaxing the region-covering constraint (C5) into the objective with multipliers $\lambda_k$ decomposes the problem into a shortest-path subproblem and could provide tighter bounds than the MTZ LP relaxation.

## 12. Conclusion

We modelled the aerodrome race problem using two ILP formulations (MTZ and SEC) and a greedy heuristic, and tested them on 10 instances.

**MTZ** proved optimal on 7 of 9 instances within 120 s. Its $O(n^2)$ constraints and single-solve approach make it the practical winner for our implementation.

**SEC** consistently explores far fewer B&B nodes (up to $376\times$ fewer), confirming the tighter LP bound. However, our iterative re-solve overhead meant SEC proved optimal on only 2 of 9 instances. The correct conclusion is: *iterative SEC is slower than lazy-callback SEC*, not that the SEC formulation is inferior.

**Heuristic** returns solutions in under 1 ms with gaps of 5–14% on feasible instances. It fails on sparse graphs due to range violations in the Dijkstra escape mechanism.

**What would improve results:**
- Implement SEC with lazy callbacks (single B&B run, no restarts)
- Use the heuristic solution as a warm-start incumbent for both ILP models
- Apply Lagrangian relaxation on constraint (C5) for a tighter lower bound without SEC overhead
